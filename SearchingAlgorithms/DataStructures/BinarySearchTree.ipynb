```python
class Node:
    def __init__(self, data):
        self.data = data
        self.right_child = None
        self.left_child = None
```

```python
class BinarySearchTree:
    def __init__(self):
        self.root_node = None

    def find_min(self):
        current = self.root_node
        while current:
            current = current.left_child

        return current

    def find_max(self):
        current = self.root_node
        while current:
            current = current.right_child

        return current

    def insert(self, data):
        new_node = Node(data)
        if self.root_node is None:
            self.root_node = new_node
        else:
            current = self.root_node
            parent = None
            while True:
                parent = current
            if new_node.data < current.data:
                current = current.left_child
                if current is None:
                    parent.left_child = new_node
                    return
            else:
                current = current.right_child
                if current is None:
                    parent.right_child = new_node
                    return
            return

    #Considering that our node class does not have a reference to a parent of a       #node's data, we will implement a helper method to search for and return the      #node with its parent node
    def get_node_with_parent(self, data):
        parent = None
        current = self.root_node
        if current is None:
            return (parent, None)
        while True:
            if current.data == data:
                return (parent, current)
            elif current.data > data:
                parent = current
                current = current.left_child
            else:
                parent = current
                current = current.right_child

        return (parent, current)
        
    #The method to do the actual removal 
    def remove(self, data):
        parent, node = get_node_with_parent(data)

        #To check the no. of children that the node we want to delete/remove has            #we will use the if statements below
        if parent is None and node is None:
            return False

        #To get the no. of children in the node we initialize a variable to hold
        children_count = 0

        if node.left_child and node.right_child:
            children_count = 2
        elif (node.left_child is None) and (node.right_child is None):
            children_count = 0
        else:
            children_count = 1

        #Next, we need to implement code that handles conditions under which a              #node can and will be deleted
        #The first part of the if statement below handles the case where the node           #has no children i.e the children_count is 0
        if children_count == 0:
            #And at least has a parent
            if parent:
                if parent.right_child is node:
                    parent.right_child = None
                else:
                    parent.left_child = None
            #If there's no parent, which means there's only the root node
            else:
                self.root_node = None
        #In the case where the node about to be deleted has one child:
        elif children_count == 1:
            #We will use next_node to keep track of where the single node pointed   #to by the node we want to delete is, we will then connect parent.left_child or     #parent.right_child to next_node
            next_node = None
            if node.left_child:
                next_node = node.left_child
            else:
                next_node = node.right_child

            if parent:
                if parent.left_child is node:
                    parent.left_child = next_node
                else:
                    parent.right_child = next_node
            else:
                self.root_node = next_node
        #In the else condition we handle when the children_count is 2
        else:
            parent_of_leftmost_node = node
            leftmost_node = node.right_child
            while leftmost_node.left_child:
                parent_of_leftmost_node = leftmost_node
                leftmost_node = leftmost_node.left_child

            node.data = leftmost_node.data

        if parent_of_leftmost_node.left_child == leftmost_node:
            parent_of_leftmost_node.left_child = leftmost_node.right_child
        else:
            parent_of_leftmost_node.right_child = leftmost_node.right_child


    def search(self, data):
        current = self.root_node
        while True:
            if current is None:
                return None
            elif current.data > data:
                current = current.left_child
            else:
                current = current.right_child

    #Tree Traversal, beginning with Depth-first methods
    #First of the Depth-first methods is the in-order traversal technique
    #For in-order, we first visit the left sub-tree, then the root node and             #finally the right sub-tree
    def in_order(self, root_node):
        current = self.root_node
        if current is None:
            return
        self.in_order(current.left_child)
        print(current.data)
        self.in_order(current.right_child)

    #For pre-order, we first visit the root_node, then the left sub-tree and            #finally the right sub-tree
    def pre_order(self, root_node):
        current = self.root_node
        if current is None:
            return
        print(current.data)
        self.pre_order(current.left_child)
        self.pre_order(current.right_child)

    #For post-order, we visit the left sub-tree, then the right sub-tree and            #finally the root_node
    def post_order(self, root_node):
        current = self.root_node
        if current is None:
            return
        self.post_order(current.left_child)
        self.post_order(current.right_child)
        print(current.data)

    #Next, we implement breadth-first traversal
    def breadth_first_traversal(self):
        list_of_nodes_visited = []
        traversal_queue = deque([self.root_node])
        while len(traversal_queue) > 0:
            node = traversal_queue.popleft()
            list_of_nodes_visited.append(node.data)

            if node.left_child:
                traversal_queue.append(node.left_child)
            
            if node.right_child:
                traversal_queue.append(node.right_child)

        return list_of_nodes_visited

                    
```

So far, we have been dealing only with nodes that contain only integers or numbers. For
numbers, the idea of greater than and lesser than are clearly defined. Strings would be
compared alphabetically, so there are no major problems there either. But if you want to
store your own custom data types inside a BST, you would have to make sure that your
class supports ordering.
